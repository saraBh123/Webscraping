```python
# On cr√©e une instance du web-driver Firefox et on va sur la page de eBay.fr
driver = webdriver.Chrome(executable_path=r"C:/Program Files (x86)/Google/Chrome/Application/chromedriver.exe")
```

```python
requete = requests.get("https://www.nhsinform.scot/illnesses-and-conditions/a-to-z")
page = requete.content
soup = BeautifulSoup(page)
```

```python
link = soup.find_all("a", {"class": "nhs-uk__az-link"})
list_link=[]
for link in soup.find_all('a', {"class": "nhs-uk__az-link"}):
    l = link.get('href')
    list_link.append("https://www.nhsinform.scot"+l)
    
```

```python
list_link
```

```python
titre = []
for l in list_link :
    driver.get(l)
    h2 = driver.find_elements(By.TAG_NAME, 'h2')
    try:
        for x in range(len(h2)):
            titre.append(h2[x].text)
    except Exception:
        pass
    
```

```python
import numpy as np
titre2 = [x for x in titre if x != '']
```

```python
desc = []
for l in list_link :
    driver.get(l)
    p = driver.find_elements(By.TAG_NAME, 'p')
    try:
        for x in range(len(p)):
            desc.append(p[x].text)
    except Exception:
        pass
```

```python
df = pd.concat([titre, desc], axis=1)
df.head()
```
